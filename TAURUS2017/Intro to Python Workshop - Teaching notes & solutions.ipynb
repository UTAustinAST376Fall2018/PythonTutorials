```python
# Exercise 3 solutions

array = np.array([ 4,  0.5,  6,  5, 11, 14, 12, 14,  7, 16])

emptylist = []

for i in range(0, len(array)):
    if array[i]**2 < 100.:
        emptylist.append(array[i])

print(emptylist)
```

```python
emptylist2 = []

for i in range(0, len(array)):
    if array[i]**2 < 100.:
        emptylist2.append([array[i], array[i]**2])
        
print(emptylist2)
#print(emptylist2.size)
emptylist2_asarray = np.asarray(emptylist2)
print(emptylist2_asarray.size)
```

```python
# Exercise 4 solutions

def gauss(mu, sig, x):
    const = (sig**2 * np.sqrt(2*np.pi))
    exp = np.exp(-(x-mu)**2/(2*sig**2))
    return const*exp

xarr = np.linspace(-5, 5, 1000)

gaussian = gauss(0, 1, xarr)

import matplotlib.pyplot as plt
plt.plot(xarr, gaussian)
plt.show()
```

```python
# Exercise 5 solutions

g1 = gauss(0, 1, xarr)
g2 = gauss(0.5, 1, xarr)
g3 = gauss(0, 0.5, xarr)

plt.plot(xarr, g1, 'g--', label='$\mu=0, \sigma=1$')
plt.plot(xarr, g2, 'k-', label='$\mu=0.5, \sigma=1$')
plt.plot(xarr, g3, 'm-.', label='$\mu=0, \sigma=0.5$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gauss')
plt.legend()
plt.show()
```

```python
# exercise 6 solutions
from scipy.stats import norm

sig2 = np.random.normal(0, 2, 10000)
sig15 = np.random.normal(0, 1.5, 10000)

plt.hist(sig2, bins=50, hatch='|', alpha=0.5, facecolor='magenta', label='$\sigma=2$', normed=True)
plt.hist(sig15, bins=50, alpha=0.25, facecolor='cyan', label='$\sigma=1.5$', normed=True)

x=np.linspace(-6, 6, 10000)
mu2, sigma2 = norm.fit(sig2)
p2 = norm.pdf(x, mu2, sigma2)
plt.plot(x, p2, 'm', linewidth=2)
mu15, sigma15 = norm.fit(sig15)
p15 = norm.pdf(x, mu15, sigma15)
plt.plot(x, p15, 'c', linewidth=1.5)

plt.legend()
plt.show()
```
